```python
import plotly.express as px
import pandas as pd
df= pd.read_csv('./경기도 수원시_지역화폐 결제 정보_20201211_수정.csv', encoding='cp949')
df.head()
```

```python
#Substr 문자열 자르기를 위해 데이터타입 변경
df= df.astype({'기준년월' : 'string'})
df.info()
```

```python
# 기준년 생성, 불필요 컬럼 삭제
df['기준년'] = df['기준년월'].str[:4]
df['기준월'] = df['기준년월'].str[5:7]
del df['데이터기준일자']
df
```

```python
# import libraries
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
# set plot style : grey grid in the backgroud;
sns.set(style="darkgrid")

# set the figure size
plt.figure(figsize=(30,30))
```

```python
df = df.set_index('기준년')
df_2019 = df.loc['2019']
df_2019
```

```python
# top bar -> sum all values
df_age = df_2019.groupby(['연령대, 업종명'])['결제금액'].sum().sort_values(ascending=False).index[:10]
df_Ind = df_2019.groupby(['업종명'])['결제금액'].sum().sort_values(ascending=False).index[:10]
# print(df_Ind)
df_2019 = df_2019.reset_index(inplace=True)
df_2019 = df_2019.set_index('업종명')
df_2019_Ind = df_2019.loc[df_Ind]
df_2019_Ind
```
