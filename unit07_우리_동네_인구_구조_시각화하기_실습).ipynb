우리 동네 연령별 인구수 result 리스트에 저장하기 -> 시각화를 위해

```python
# 리스트 내포를 통해 할 수도 있음

import csv

f=open('age.csv',encoding='cp949')
data = csv.reader(f)

result=[]

for row in data:
  if '철산3동' in row[0] :
    for i in row[3:]:
      result.append(int(i)) #인구수 데이터 정수로 변환 후 result 변수에 담기

import csv

f=open('cheolsan.csv',encoding='cp949')
data = csv.reader(f)
result2=[]

for row in data:
  if '철산3동' in row[0] :
    for i in row[3:]:
      result2.append(int(i.replace(",",""))) #인구수 데이터 정수로 변환 후 result 변수에 담기

print(result2)
print(result)
```
