```python
#라이브러리 사용
import tensorflow as tf
import pandas as pd
```

```python
# 데이터 준비
url = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/boston.csv'
bt = pd.read_csv(url)

print(bt.columns)
```

```python
# 종속변수, 독립변수
cause = bt[['crim', 'zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax',
       'ptratio', 'b', 'lstat']]
effect = bt[['medv']]
print(cause.shape, effect.shape)
```

```python
# 2. 모델 구조 만들기
X = tf.keras.layers.Input(shape=[13])
H = tf.keras.layers.Dense(10, activation='swish')(X) #X인풋을 받는다
Y = tf.keras.layers.Dense(1)(H)
model = tf.keras.models.Model(X,Y)
model.compile(loss = 'mse')
```

```python
# 모델 구조 확인
model.summary() # 13개입력받았음, 2번째-> 10개입력받아서 1개layer받는모델추가
# dense_2 항이 10개 b가 1개
# inputLayer 14개 가중치가 필요, dense_2 -> 14*10개 140개 가중치
```

```python
# 3. 데이터로 모델 학습(FIT)
model.fit(cause, effect, epochs=1000, verbose=0)
model.fit(cause, effect, epochs=100)
```

```python
# 4. 모델 이용
print(model.predict(cause[:5]))
print(effect[:5])
```

```python
# 데이터 준비
url_iris = 'https://raw.githubusercontent.com/blackdew/tensorflow1/master/csv/iris.csv'
ir = pd.read_csv(url_iris)

#원핫인코딩
ir = pd.get_dummies(ir)

ir.columns
```

```python
# 종속변수, 독립변수 나누기
cause = ir[['꽃잎길이', '꽃잎폭', '꽃받침길이', '꽃받침폭']]
effect = ir[['품종_setosa', '품종_versicolor',
       '품종_virginica']]
print(cause.shape, effect.shape)
```

```python
#2. 모델의 구조 만들기
X = tf.keras.layers.Input(shape=[4])
H = tf.keras.layers.Dense(8, activation="swish")(X)
H = tf.keras.layers.Dense(8, activation="swish")(H)
H = tf.keras.layers.Dense(8, activation="swish")(H)
H = tf.keras.layers.Dense(3, activation="swish")(H)
Y = tf.keras.layers.Dense(3, activation='softmax')(H)
model = tf.keras.models.Model(X,Y)
model.compile(loss='categorical_crossentropy', metrics='accuracy')
```

```python
#모델 구조 확인
model.summary()
```

```python
#3. 데이터 모델 학습(FIT)
model.fit(cause, effect, epochs=100)
```

```python
#4. 모델을 이용
print(model.predict(cause[0:5]))
print(effect[0:5])
#9.99 * 10((-1)) -> 1에 가까운 숫자
```
