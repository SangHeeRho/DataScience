꺾은선 그래프로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f=open('gender.csv',encoding='cp949')
data = csv.reader(f)

m=[]
f=[]

#지역명을 입력받는 변수 선언
name = input("찾고 싶은 지역의 이름을 알려주세요 : ")

for row in data :
  if name in row[0]:
    for i in row[3:104] :# 남성 데이터 :3 - 103번 데이터 슬라이싱 - 정수 데이터로 변환 - m 리스트에 저장
      m.append(int(i.replace(',','')))
    for i in row[106:]: # 여성 데이터 :106번 - 끝까지 데이터 슬라이싱 - 정수 데이터로 변환 - f 리스트에 저장
      f.append(int(i.replace(',','')))
    break
plt.style.use('ggplot')
plt.figure(dpi=200)
plt.rc('font',family='NanumBarunGothic')

# 꺾은선 그래프로 남여 인구수 데이터 표현하기
plt.plot(m,label='남성')
plt.plot(f, label='여성')
plt.legend()
plt.show()
```

연령별로 남성 인구수에서 여성 인구수를 뺀 차이의 막대그래프 그리기

```python
import csv
import matplotlib.pyplot as plt

f=open('gender.csv',encoding='cp949')
data = csv.reader(f)

result =[] # 남성 인구 - 여성 인구 값 담을 리스트 변수 선언

name = input("찾고 싶은 지역의 이름을 알려주세요 : ")
for row in data :
  if name in row[0]:
    for i in range(3,104):
      result.append(int(row[i].replace(',','')) - int(row[i+103].replace(',','')))
    break

plt.style.use('ggplot')
plt.figure(dpi=200)
plt.rc('font',family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title(name+ ' 지역의 남녀 성별 인구 차이')
plt.bar(range(101),result)
plt.show()
```

위치, 크기가 서로 다른 100개의 점 만들기

```python
import matplotlib.pyplot as plt
import random

x=[]
y=[]
size = []

for i in range(100):
  x.append(random.randint(50,100))
  y.append(random.randint(50,100))
  size.append(random.randint(10,100))

plt.figure(dpi=200)
plt.style.use('ggplot')
plt.scatter(x,y, s=size)
plt.show()
```

```python
plt.figure(dpi=200)
plt.style.use('ggplot')
plt.scatter(x,y, s=size, c=size, cmap='jet')
plt.colorbar()
plt.show()
```

```python
plt.figure(dpi=200)
plt.style.use('ggplot')
plt.scatter(x,y, s=size, c=size, cmap='jet', alpha=0.7) # alpha 투명도 추가 0~1: 0에 가까울수록 투명, 1에 가까울수록 불투명
plt.colorbar()
plt.show()
```

제주도의 연령대별 성별 비율을 산점도로 표현하기

남녀 연령별 인구 데이터 scatter() 함수로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f=open('gender.csv',encoding='cp949')
data = csv.reader(f)

m=[]
f=[]

#지역명을 입력받는 변수 선언
name = input("찾고 싶은 지역의 이름을 알려주세요 : ")

for row in data :
  if name in row[0]:
    for i in row[3:104] :# 남성 데이터 :3 - 103번 데이터 슬라이싱 - 정수 데이터로 변환 - m 리스트에 저장
      m.append(int(i.replace(',','')))
    for i in row[106:]: # 여성 데이터 :106번 - 끝까지 데이터 슬라이싱 - 정수 데이터로 변환 - f 리스트에 저장
      f.append(int(i.replace(',','')))
    break
plt.style.use('ggplot')
plt.figure(dpi=200)
plt.rc('font',family='NanumBarunGothic')

#산점도 그래프로 남여 인구수 데이터 표현하기
plt.scatter(m,f) # x축은 남자, y축은 여자
plt.show()
```

원의 크기는 인구수의 제곱근 추가

```python
import csv
import matplotlib.pyplot as plt
import math

f=open('gender.csv',encoding='cp949')
data = csv.reader(f)

m=[]
f=[]
size=[] # 인구수 사이즈 리스트 변수 선언

#지역명을 입력받는 변수 선언
name = input("찾고 싶은 지역의 이름을 알려주세요 : ")

for row in data :
  if name in row[0]:
    for i in row[3:104] :
      m.append(int(i.replace(',','')))
      size.append(math.sqrt(int(i.replace(',',''))))
    for i in row[106:]:
      f.append(int(i.replace(',','')))
    break

plt.style.use('ggplot')
plt.rc('font',family='NanumBarunGothic')

#인구수와 sqrt(인구수) 차이 확인 (데이터가 너무 크면 sqrt 효과를 줌)
print(m)
print(size)

#산점도 그래프로 남여 인구수 데이터 표현하기
plt.figure(dpi=200)
plt.scatter(m,f,s=size,c=range(101),cmap='jet', alpha=0.5) # x축은 남자, y축은 여자
plt.colorbar()
plt.plot(range(max(m)),range(max(m)),'g') #추세선(기준) 추가
plt.show()
```
