```python
import matplotlib.pyplot as plt # 1) 라이브러리 임포트

plt.title('linestyle') # 타이틀 속성 추가

plt.plot([10,20,30,40], color='r', linestyle = '--', label='dashed') # 기본 색상은 약어로 표시 가능
plt.plot([40,30,20,10], 'g', ls= ':', label='dotted') # 초록색 dotted 그래프 linestyle -> ls 표현 가능

plt.legend() # 그래프에 범례 속성 추가
plt.show()
```

```python
import matplotlib.pyplot as plt # 1) 라이브러리 임포트

plt.title('marker') # 타이틀 속성 추가

# 'r.--' : <색상><마커모양><선모양> 코드로 작성 가능
plt.plot([10,20,30,40], 'r.', label='circle') # 빨간색 circle 그래프
plt.plot([40,30,20,10], 'g^', label='triangle up') # 초록색 triangle up 그래프 linestyle -> ls 표현 가능

plt.legend() # 그래프에 범례 속성 추가
plt.show()
```

데이터 시각화하기

```python
import csv
import matplotlib.pyplot as plt # 1) 라이브러리 임포트

f=open('seoul.csv', encoding='cp949')
data = csv.reader(f)
header = next(data)

result=[] # 결측치를 제거한 최고 기온 리스트

for row in data :
  if row[-1] != '' : # 최고 기온이 ''(빈문자열)이 아니면,
    result.append(float(row[-1]))

result

plt.figure(figsize=(10,2), dpi=200)
plt.plot(result, 'r')
plt.show()
```

```python
import csv

min_temp = 999 #최저 기온 값을 저장할 변수 -> 명확한 결과를 위해 극적인 값으로 셋팅
min_date = '' # 최저 기온이 가장 낮았던 날짜를 저장하는 변수

f = open('/content/seoul_2017-2020.csv','r',encoding='cp949')

data = csv.reader(f)
header = next(data) # 헤더 추출

for row in data:

  #최저 기온 데이터가 ''(결측치)인지 확인
  if row[-2] == '':
    row[-2] = 999 # 999를 넣어서 빈 문자열('')이 있던 자리라고 표시

  row[-2] = float(row[-2]) # 기온 데이터 실수형으로 변환

  #최저 기온 정보를 저장

  if min_temp > row[-2] :
    min_date = row[0]
    min_temp = row[-2]

f.close()

print('기상 관측 이래 서울의 최저 기온이 가장 낮았던 날은', min_date+'로 ', min_temp, '도 였습니다.')
```
