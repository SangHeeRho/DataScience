히스토그램 hist() 함수 사용

```python
import matplotlib.pyplot as plt

plt.figure(dpi=100) # 해상도
# 히스토그램 -> 도수 분포 -> 어떠한 자료에 속해 있는 값들을 수량으로 나타낸 것
plt.hist([1,1,2,3,4,5,6,7,8,10]) # hist(data) 가로축은 계급(item), 세로축은 도수(횟수, 빈도수)
plt.show()
```

주사위 시뮬레이션-히스토그램 연습

```python
import random
print(random.randint(1,6)) #1~6 사이의 값을 랜덤으로 출력
```

```python
# 주사위 5번 던져서 나온 데이터 출력

import random

for i in range(5):
  print(random.randint(1,6))
```

```python
# 주사위 5번 던져서 나온 데이터 최종 한번에 결과로 한번에 출력

import random
import matplotlib.pyplot as plt

dice = [] # 주사위 리스트 변수 선언

for i in range(5):
  dice.append(random.randint(1,6))

#랜덤하게 굴린 주사위 5개의 값을 히스토그램으로 표현
plt.figure(dpi=200)
plt.hist(dice,bins=6)
plt.show()
```

최고 기온 데이터 상자 그림으로 표현하기

1월부터 12월까지 최고 기온 데이터 상자 그림으로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv',encoding='cp949')
data=csv.reader(f)
next(data)

month=[[],[],[],[],[],[],[],[],[],[],[],[]] #월 별 데이터 저장할 month 변수 선언

for row in data:
  if row[-1] != '':
    month[int(row[0].split('-')[1])-1].append(float(row[-1])) # 월과 같은 번호의 인덱스에 월별 데이터 저장 (예: 1월 row[0].split('-')[1] -> month[0] row.split('-')[1]-1, 2월 -> month[1])

plt.figure(dpi=200)
plt.boxplot(month)
plt.show()
```

8월 1일부터 31일까지 최고 기온 데이터 상자 그림으로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv',encoding='cp949')
data=csv.reader(f)
next(data)

day=[] #일별 데이터를 저장할 리스트 day 변수 선언
for i in range(31) :
  day.append([]) # day 리스트내에 31개의 리스트 생성[[],[],,,,,31개]

for row in data:
  if row[-1] !='' :
    if row[0].split('-')[1] == '08' : #8월 최고 기온 값 저장 day[0] -> 1일
      day[int(row[0].split('-')[2])-1].append(float(row[-1]))             
plt.style.use('ggplot') # 그래프 스타일 지정 (회색 배경의 격자무늬, 2/4 값을 의미하는 선의 색 변경)
plt.figure(figsize=(10,5),dpi=200)
plt.boxplot(day, showfliers=False) #아웃라이어(이상치) 값 생략
plt.show()
```
