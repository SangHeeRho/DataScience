```python
# 최고기온 값이 8월에 해당되는 값만 추출해서 그래프 시각화

import csv
import matplotlib.pyplot as plt

f=open('seoul.csv',encoding='cp949')
data=csv.reader(f)
next(data)

result=[] # 최고 기온 데이터 중 8월 데이터만 저장할 리스트 생성

for row in data:
  if row[-1] !='': # 최고 기온 결측치 제외
    if row[0].split('-')[1] == '08' : # 날짜를 잘라서 08 월 데이터 추출
      result.append(float(row[-1])) # 결과 리스트 담기

plt.figure(dpi=100)
plt.plot(result,'hotpink') # result 리스트의 최고 기온 데이터를 빨간색 그래프로 그리기
plt.show()
```

매년 생일 기준 최고 기온 그래프 그리기

```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv',encoding='cp949')
data=csv.reader(f)
next(data)

result=[] # 최고 기온 데이터 중 8월 데이터만 저장할 리스트 생성

for row in data:
  if row[-1] !='': # 최고 기온 결측치 제외
    if row[0].split('-')[1] == '12' and row[0].split('-')[2] == '24' : # 날짜를 잘라서 08 월 데이터 추출
      result.append(float(row[-1])) # 결과 리스트 담기

plt.figure(dpi=100)
plt.plot(result,'hotpink') # result 리스트의 최고 기온 데이터를 빨간색 그래프로 그리기
plt.show()
```

매년 생일 기준 최고 기온, 최저 기온 그래프 그리기

```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv',encoding='cp949')
data=csv.reader(f)
next(data) # 헤더

high =[] # 최고 기온
low =[] # 최저 기온

for row in data:
  if row[-1] != '' and row[-2] != '':
    if row[0].split('-')[1] == '08' and row[0].split('-')[2] == '14':
      high.append(float(row[-1])) # 최고 기온
      low.append(float(row[-2])) # 최저 기온

plt.figure(dpi=100)
plt.plot(high, 'hotpink', label = 'high') # 최고 기온
plt.plot(low, 'skyblue', label = 'low') # 최저 기온
plt.legend()
plt.show()
```

매년 생일 기준 최고 기온, 최저 기온, 평균 기온 그래프 그리기

```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv',encoding='cp949')
data=csv.reader(f)
next(data) # 헤더

high = [] # 최고 기온
avg = [] # 평균 기온
low = [] # 최저 기온


for row in data:
  if row[-1] != '' and row[-2] != '' and row[-3] != '':
    if 1994 <= int(row[0].split('-')[0]):
      if row[0].split('-')[1] == '09' and row[0].split('-')[2] == '14':
        high.append(float(row[-1]))
        low.append(float(row[-2]))
        avg.append(float(row[-3]))


#데이터 시각화
#차트 설정
plt.figure(dpi=100) # 해상도
plt.rc('font', family = 'NanumBarunGothic') # 나눔 바른 고딕을 기본 글꼴로 설정
plt.rcParams['axes.unicode_minus'] = False # 마이너스 기호 깨짐 방지
#plot 차트 그리기
plt.plot(high, 'hotpink', label='최고 기온')
plt.plot(low, 'skyblue', label='최저 기온')
plt.plot(avg, 'g', label='평균 기온')

plt.title("SangHee's birthday Temp")
plt.legend()

plt.show()
```

1워로가 8월의 데이터를 히스토그램으로 시각화하기

```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv',encoding='cp949')
data=csv.reader(f)
next(data)

aug = [] # 8월의 최고 기온을 담을 aug 리스트 변수 선언
jan = [] # 1월의 최ㅣ고 기온을 담을 jan 리스트 변수 선언

for row in data:
  month = row[0].split('-')[1]
  if row[-1] != '' :
    if month == '08' : # 8월 데이터면 aug 변수에 담기
      aug.append(float(row[-1]))
    if month == '01' : # 1월 데이터면 jan 변수에 담기
      jan.append(float(row[-1]))

#최고 기온 데이터 히스토그램
plt.figure(dpi=200)
plt.hist(aug,bins=100,color='r', label="Aug")
plt.hist(jan,bins=100,color='b', label="Jan")
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f=open('seoul.csv',encoding='cp949')
data=csv.reader(f)
next(data)

aug = [] # 8월의 최고 기온을 담을 aug 리스트 변수 선언
jan = [] # 1월의 최ㅣ고 기온을 담을 jan 리스트 변수 선언
june = [] # 6월의 최고 기온을 담을 june 리스트 변수 선언

for row in data:
  month = row[0].split('-')[1]
  if row[-1] != '' :
    if month == '08' : # 8월 데이터면 aug 변수에 담기
      aug.append(float(row[-1]))
    if month == '01' : # 1월 데이터면 jan 변수에 담기
      jan.append(float(row[-1]))
    if month == '06' : # 6월 데이터면 jan 변수에 담기
      june.append(float(row[-1]))

#최고 기온 데이터 히스토그램
plt.figure(dpi=20)
plt.hist(aug,bins=100,color='r', label="Aug")
plt.hist(jan,bins=100,color='b', label="Jan")
plt.hist(june,bins=100,color='g', label="Jan")
plt.legend()
plt.show()
```
